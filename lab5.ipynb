```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
x_rand = np.linspace(-10, 10, num = 500) # for generating random samples
x_rand
```

```python
def function(x):
    return (0.5*(x**3))-(x**2)+2 + np.random.normal(-10,50,500)  #Generating samples using this function
y = function(x_rand)
```

# Visualizing the function

```python
plt.scatter(x_rand, y, c = 'b', marker = 'x') #using 
plt.title("Visualizing the function")
plt.show()
```

```python
linear_model = LinearRegression()
X = x_rand.reshape(-1, 1) #Converted the dataset into convinient shape
linear_model.fit(X, y)
# model.predict(x_rand)
```

```python
y_pred = linear_model.predict(X)
y_pred
```

```python
plt.scatter(X, y, c = 'b', marker = 'x')
plt.title("Fitting a linear graph")
plt.plot(X, y_pred, c = 'r')
plt.show()
```

```python
poly = PolynomialFeatures(degree = 3)
X2_poly = poly.fit_transform(X)

model_3 = LinearRegression()
model_3.fit(X2_poly, y)
```

```python
y_pred_3 = model_3.predict(X2_poly)
plt.scatter(X, y, c = 'b', marker = 'x')
plt.title("Fitting a degree 3 graph")
plt.plot(X, y_pred_3, c = 'r')
plt.show()
```

```python
poly_13 = PolynomialFeatures(degree = 13)
X13_poly = poly_13.fit_transform(X)

model_13 = LinearRegression()
model_13.fit(X13_poly, y)
```

```python
y_pred_13 = model_13.predict(X13_poly)
plt.scatter(X, y, c = 'b', marker = 'x')
plt.title("Fitting a degree 13 graph")
plt.plot(X, y_pred_13, c = 'r')
plt.show()
```

# Metrics 

For linear model

```python
#MSE
mse_linear = mean_squared_error(y, y_pred)
print("Mean squared error of linear model: ", mse_linear)
#MAE
mae_linear = mean_absolute_error(y, y_pred)
print("Mean absolute error of linear model: ", mae_linear)
#r2score
r2_linear = r2_score(y, y_pred)
print("r2 score of linear model: ", r2_linear)
```

For degree 3 polynomial

```python
#MSE
mse_3 = mean_squared_error(y, y_pred_3)
print("Mean squared error of degree 3 model: ", mse_3)
#MAE
mae_3 = mean_absolute_error(y, y_pred_3)
print("Mean absolute error of degree 3 model: ", mae_3)
#r2score
r2_3 = r2_score(y, y_pred_3)
print("r2 score of degree 3 model: ", r2_3)
```

For Degree 13

```python
#MSE
mse_13 = mean_squared_error(y, y_pred_13)
print("Mean squared error of degree 13 model: ", mse_13)
#MAE
mae_13 = mean_absolute_error(y, y_pred_13)
print("Mean absolute error of degree 13 model: ", mae_13)
#r2score
r2_13 = r2_score(y, y_pred_13)
print("r2 score of degree 13 model: ", r2_13)
```

By analysing the three fits we can say that degree 13 is the best fit
