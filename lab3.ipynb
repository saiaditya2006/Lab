```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
housing = fetch_california_housing()
print(housing.data.shape, housing.target.shape)
```

```python
print(housing.feature_names[:])
```

```python
print(housing.data)
```

```python
df = pd.DataFrame(housing.data, columns = housing.feature_names)
df.head()
```

```python
df['Target'] = housing.target
df.head()
```

```python
X_train = df.drop(columns=['Target'])
y_train = df['Target']
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

# Task 1

```python
multiple_linear_model = LinearRegression()
multiple_linear_model.fit(X_train, y_train)
```

```python
y_pred = multiple_linear_model.predict(X_train)
```

```python
mse_multiple = mean_squared_error(y_train, y_pred)
print("The MSE with multiple linear regression is: ", mse_multiple)
```

```python
r2_score_multiple = r2_score(y_train, y_pred)
print("The r2 score with multiple linear regression is: ", r2_score_multiple)
```

# Task 2

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(X_poly, y_train)
```

```python
y2_pred = model.predict(X_poly)
y2_pred
```

```python
mse_sklearn = mean_squared_error(y_train, y2_pred)
print("Sklearn MSE:", mse_sklearn)
```

```python
r2_score_poly = r2_score(y_train, y2_pred)
print("The r2 score with multiple linear regression is: ", r2_score_poly)
```
