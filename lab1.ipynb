```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = np.array([[2,3], [-4, 1]])
B = np.array([6, -8])

A_inv = np.linalg.inv(A)
solution = np.dot(A_inv, B)
x, y =  solution

print(f"The solution of x = {x}, y = {y}")

# or
solution1 = np.linalg.solve(A, B)
x1, y1 = solution1
print(f'The solution of x = {x1}, y = {y1}')
```

```python
# Visualization
x_vals = np.linspace(-10, 10, 100)
y1_vals = (6 - 2*x_vals) / 3
y2_vals = (-8 + 4*x_vals) / 1 
```

# Solving and Visualizing in 2d system

```python
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label='2x + 3y = 6', color='blue')
plt.plot(x_vals, y2_vals, label='-4x + y = -8', color='red')
plt.scatter(x, y, color='black', marker='o', label='Intersection')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("Intersection of Two Lines (2D)")
plt.grid()
plt.show()
```

```python
A = np.array([[2,3,1], [-4, 1, 2], [1, -2, 3]])
B = np.array([6, -8, 4])

A_inv = np.linalg.inv(A)
solution = np.dot(A_inv, B)
x, y, z =  solution

print(f"The solution of x = {x}, y = {y}, z = {z}")

# or
solution1 = np.linalg.solve(A, B)
x1, y1, z1 = solution1
print(f'The solution of x = {x1}, y = {y1}, z = {z1}')
```

```python
x_vals = np.linspace(-10, 10, 10)
y_vals = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x_vals, y_vals)

Z1 = (6 - 2*X - 3*Y) 
Z2 = (-8 + 4*X - Y) / 2 
Z3 = (4 - X + 2*Y) / 3

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.5, color='blue', label="Plane 1")
ax.plot_surface(X, Y, Z2, alpha=0.5, color='red', label="Plane 2")
ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', label="Plane 3")

ax.scatter(x, y, z, color='black', s=100, label="Intersection")

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Intersection of Three Planes (3D)")
plt.legend()
plt.show()
```

```python
def check_solution(A, B):
    rank_A = np.linalg.matrix_rank(A)
    rank_Aug = np.linalg.matrix_rank(np.column_stack((A, B)))

    if rank_A == rank_Aug:
        if rank_A == A.shape[1]:  
            return "Unique Solution"
        else:
            return "Infinite Solutions"
    else:
        return "No Solution"

# Example cases:
A_inf = np.array([[1, -1, 2], [2, -2, 4], [3, -3, 6]]) 
B_inf = np.array([3, 6, 9])

A_no_sol = np.array([[1, -1, 2],
              [2, -2, 4],
              [3, -3, 6]])
B_no_sol = np.array([1, 2, 10])

print(f"Case 1 (Unique Solution): {check_solution(A, B)}")
print(f"Case 2 (Infinite Solutions): {check_solution(A_inf, B_inf)}")
print(f"Case 3 (No Solution): {check_solution(A_no_sol, B_no_sol)}")
```

# Repeating the same for gaussian distribution

```python
#Home Work
# A) Create a plot of 10000 points generated from 2-D gaussian  distribution and calculate mean, median.
# B) Add few outliers outside the range of these points and calculate mean, median.

mean = [0,0]
cov = [[1,0], [0,1]]

points = np.random.multivariate_normal(mean, cov, 10000)

mean_original = np.mean(points, axis = 0)
median_original = np.median(points, axis = 0)

plt.scatter(points[:,0], points[:,1], s = 1, label = 'Original Points')
plt.title('2D Gaussian Distribution (10000 points)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

```

# Adding Outliers

```python
outliers = np.array([[10,10], [12, -12], [-15,15], [-20, 20], [25,25]])
points_with_outliers = np.vstack((points, outliers))

mean_with_outliers = np.mean(points_with_outliers, axis = 0)
median_with_outliers = np.median(points_with_outliers, axis =0)

plt.scatter(points_with_outliers[:,0], points_with_outliers[:,1], s=1, label = 'Points With Outliers')
plt.scatter(outliers[:,0], outliers[:,1], color = 'red', label = 'Outliers', marker = 'x')
plt.title('2D Gaussian Distribution with Outliers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

(mean_original, median_original), (mean_with_outliers, median_with_outliers)
```

# Uniform Distribution

```python
np.random.seed(42)  
x = np.random.uniform(-10, 10, 10000)
y = np.random.uniform(-10, 10, 10000)

mean_x, mean_y = np.mean(x), np.mean(y)
median_x, median_y = np.median(x), np.median(y)

print("Before adding outliers:")
print(f"Mean: ({mean_x:.2f}, {mean_y:.2f})")
print(f"Median: ({median_x:.2f}, {median_y:.2f})")

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=1, alpha=0.5, label="Original Data")
plt.axvline(mean_x, color='red', linestyle='--', label="Mean X")
plt.axhline(mean_y, color='blue', linestyle='--', label="Mean Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter Plot of 10,000 Points (Before Outliers)")
plt.show()
```

# Adding Outliers

```python
outliers_x = np.array([50, -50, 60, -60])
outliers_y = np.array([50, -50, 60, -60])
x_with_outliers = np.concatenate((x, outliers_x))
y_with_outliers = np.concatenate((y, outliers_y))

mean_x_out, mean_y_out = np.mean(x_with_outliers), np.mean(y_with_outliers)
median_x_out, median_y_out = np.median(x_with_outliers), np.median(y_with_outliers)

print("\nAfter adding outliers:")
print(f"Mean: ({mean_x_out:.2f}, {mean_y_out:.2f})")
print(f"Median: ({median_x_out:.2f}, {median_y_out:.2f})")

plt.figure(figsize=(8, 6))
plt.scatter(x_with_outliers, y_with_outliers, s=1, alpha=0.5, label="Data with Outliers")
plt.scatter(outliers_x, outliers_y, color='red', s=50, marker='x', label="Outliers")
plt.axvline(mean_x_out, color='red', linestyle='--', label="Mean X (After Outliers)")
plt.axhline(mean_y_out, color='blue', linestyle='--', label="Mean Y (After Outliers)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter Plot of Points (After Adding Outliers)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 10,000 points from a 2D Gaussian distribution
mean_gaussian = [0, 0]  # Mean of the Gaussian distribution
cov_gaussian = [[1, 0], [0, 1]]  # Covariance matrix
data_gaussian = np.random.multivariate_normal(mean_gaussian, cov_gaussian, 10000)

# Step 2: Calculate initial mean and median
mean_initial = np.mean(data_gaussian, axis=0)
median_initial = np.median(data_gaussian, axis=0)

# Step 3: Introduce outliers
outliers = np.array([[10, 10], [12, 15], [15, -10], [-20, 20]])
data_with_outliers = np.vstack([data_gaussian, outliers])

# Step 4: Calculate new mean and median
mean_with_outliers = np.mean(data_with_outliers, axis=0)
median_with_outliers = np.median(data_with_outliers, axis=0)

# Step 5: Visualization
plt.figure(figsize=(10, 6))
plt.scatter(data_gaussian[:, 0], data_gaussian[:, 1], alpha=0.5, label='Gaussian Points')
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', label='Outliers', s=100)
plt.title("Gaussian Distribution with Outliers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Print Results
print("Without Outliers:")
print(f"Mean: {mean_initial}")
print(f"Median: {median_initial}")

print("\nWith Outliers:")
print(f"Mean: {mean_with_outliers}")
print(f"Median: {median_with_outliers}")
```
